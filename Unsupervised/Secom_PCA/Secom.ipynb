```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("./Data/secom.data",delimiter=" ",header=None)
print(data.shape)
```

```python
data.describe()
```

## Data Preprocessing

#### Replacing NULL values with median


```python
imputed_data=data.fillna(data.median())
```

```python
imputed_data.describe()
```



#### Ignoring columns with only one unique value after imputation mean as they do not contribute anything


```python
cols = data.nunique()==1
```

```python
pruned_data=imputed_data.drop(list(data.columns[cols]),axis=1)
```

```python
pruned_data.shape
```

# Principle Component Analysis (PCA)

## Step1 : Standardization

```python
from sklearn.preprocessing import StandardScaler
std_data=StandardScaler().fit_transform(pruned_data)
```

```python
std_data.shape
```

```python
std_data
```

## Step2 : Finding Covariance Matrix

```python
covar_matrix = np.matmul(std_data.T,std_data)
```

```python
covar_matrix.shape
```

## Step3 : Finding Eigen Values and Eigen Vectors

```python
from scipy.linalg import eigh
```

```python
values,vectors = eigh(covar_matrix)
```

```python
vectors.shape
```

```python
values
```

## PCA Using sklearn

```python
from sklearn import decomposition
pca = decomposition.PCA(std_data)
```

```python
pca.n_components = 474
pca_data = pca.fit_transform(std_data)
```

```python
pca.explained_variance_ 
```

### Amount of variance that is preserved across each principal component

```python
each_pc_explained_variance_ratio=pca.explained_variance_ratio_
each_pc_explained_variance_ratio
```

```python
cumulativeVariance = pca.explained_variance_ratio_.cumsum()
```

### We can see that We should consider 200 - 230, approx optimal number of principal components Principal Components to explain more than 98% of variance using below graph


```python
plt.figure(1, figsize=(6, 4))
plt.clf()
plt.plot(cumulativeVariance, linewidth=2)
plt.axis('tight')
plt.grid(True)
plt.title('Pulsar Dataset')
plt.xlabel('Number of components')
plt.ylabel('cumulative Variance')
plt.show()
```

```python
pca.n_components=7
principal7Components = pca.fit_transform(std_data)
principalDF = pd.DataFrame(data=principal7Components,columns=['pc1','pc2','pc3','pc4','pc5','pc6','pc7'])
```
